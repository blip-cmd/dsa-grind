```python
arr = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

m, n = len(arr), len(arr[0])
result = []

for d in range(m + n - 1):
    if d % 2 == 0:
        # Even diagonals: go up
        r = min(d, m - 1)
        c = d - r
        while r >= 0 and c < n:
            result.append(arr[r][c])
            r -= 1
            c += 1
    else:
        # Odd diagonals: go down
        c = min(d, n - 1)
        r = d - c
        while c >= 0 and r < m:
            result.append(arr[r][c])
            r += 1
            c -= 1

print(result)
```

```python
import random, time

nums = [random.randint(1, 10**9) for _ in range(10**6)]

# 1st Approach: string reversal
def approach1(nums):
    result = nums.copy()
    for x in nums:
        new_x = int(str(x)[::-1])
        result.append(new_x)
    return len(set(result))

# 2nd Approach: arithmetic reversal
def approach2(nums):
    seen = set(nums)
    for x in nums:
        rev = 0
        temp = x
        while temp > 0:
            digit = temp % 10
            rev = rev * 10 + digit
            temp //= 10
        seen.add(rev)
    return len(seen)

# Timing
start = time.time()
approach1(nums)
print("Approach 1:", round(time.time() - start, 2), "seconds")

start = time.time()
approach2(nums)
print("Approach 2:", round(time.time() - start, 2), "seconds")
```
